```python
import pandas as pd
df = pd.read_csv('Coursera.csv')
df.head()
```

```python
df['S.NO.'] = df.reset_index().index
df.head()
```

```python
Fdf = df.drop(['course_organization','course_Certificate_type','Total_Rating'],axis='columns')
Fdf.head()
```

```python
udf = Fdf[['S.NO.','course_title','course_rating','course_difficulty','Course_Enrollment']]
udf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
vector_array = cv.fit_transform(udf['course_title']).toarray()
```

```python
vector_array
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(vector_array)
```

```python
similarity_matrix
```

```python
def CourseRecommend(course_name):
    indexofCourseTitle = udf[udf['course_title'] == course_name].index[0]
    closeness = sorted(list(enumerate(similarity_matrix[indexofCourseTitle])),reverse=True,key = lambda x: x[1])
    for i in closeness[1:6]:
        print("The recommended Course is ",udf.iloc[i[0]].course_title)
        print("The rating for the course is %s and is subscribed by %sk users!"%(udf.iloc[i[0]].course_rating,udf.iloc[i[0]].Course_Enrollment))
        print("The difficulty for the course is %s\n"%(udf.iloc[i[0]].course_difficulty))
```

```python
CourseRecommend('Machine Learning')
```

```python
CourseRecommend('Data Science')
```
